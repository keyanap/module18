```python
# import dependencies
import pandas as pd
from datetime import datetime
```

```python
# Merge 2022 and 2023 csv files
bike_df = pd.concat(
    map(pd.read_csv, ['data/202201.csv', 'data/202202.csv', 'data/202203.csv', 'data/202204.csv', 'data/202205.csv', 'data/202206.csv', 'data/202207.csv', 'data/202208.csv', 'data/202209.csv', 'data/202210.csv', 'data/202211.csv', 'data/202212.csv', 'data/202301.csv', 'data/202302.csv', 'data/202303.csv', 'data/202304.csv', 'data/202305.csv', 'data/202306.csv', 'data/202307.csv', 'data/202308.csv', 'data/202309.csv', 'data/202310.csv', 'data/202311.csv', 'data/202312.csv']), ignore_index=True)

# Display dataframe
bike_df.head
```

```python
# Rename columns
bike_df = bike_df.rename(columns={'ride_id': 'Ride ID'})
bike_df = bike_df.rename(columns={'rideable_type': 'Rideable Type'})
bike_df = bike_df.rename(columns={'started_at': 'Started At'})
bike_df = bike_df.rename(columns={'ended_at': 'Ended At'})
bike_df = bike_df.rename(columns={'start_station_name': 'Start Station Name'})
bike_df = bike_df.rename(columns={'start_station_id': 'Start Station ID'})
bike_df = bike_df.rename(columns={'end_station_name': 'End Station Name'})
bike_df = bike_df.rename(columns={'end_station_id': 'End Station ID'})
bike_df = bike_df.rename(columns={'start_lat': 'Latitude'})
bike_df = bike_df.rename(columns={'start_lng': 'Longitude'})
bike_df = bike_df.rename(columns={'end_lat': 'End Lat'})
bike_df = bike_df.rename(columns={'end_lng': 'End Lng'})
bike_df = bike_df.rename(columns={'member_casual': 'User Type'})

# Display dataframe
bike_df.head()
```

```python
# Replace words
bike_df['Rideable Type'] = bike_df['Rideable Type'].replace(
    {'classic_bike': 'Classic Bike', 'electric_bike': 'Electric Bike', 'docked_bike': 'Docked Bike'})
bike_df['User Type'] = bike_df['User Type'].replace(
    {'member': 'Member', 'casual': 'Casual'})

# Display dataframe
bike_df.head()
```

```python
# Make starttime and stoptime datetime
bike_df['Started At'] = pd.to_datetime(bike_df['Started At'])
bike_df['Ended At'] = pd.to_datetime(bike_df['Ended At'])

# Display dataframe
bike_df.head()
```

```python
# Create start_date and start_time columns
bike_df['Start Date'] = pd.to_datetime(bike_df['Started At']).dt.date
bike_df['Start Time'] = pd.to_datetime(bike_df['Started At']).dt.time

# Create stop_date and stop_time columns
bike_df['End Date'] = pd.to_datetime(bike_df['Ended At']).dt.date
bike_df['End Time'] = pd.to_datetime(bike_df['Ended At']).dt.time

# Display dataframe
bike_df.head()
```

```python
# Make start_date and stop_date datetime
bike_df['Start Date'] = pd.to_datetime(bike_df['Start Date'])
bike_df['End Date'] = pd.to_datetime(bike_df['End Date'])
```

```python
# Convert types
bike_df['Ride ID'] = bike_df['Ride ID'].astype(str)
bike_df['Rideable Type'] = bike_df['Rideable Type'].astype(str)
bike_df['Start Station Name'] = bike_df['Start Station Name'].astype(str)
bike_df['Start Station ID'] = bike_df['Start Station ID'].astype(str)
bike_df['End Station Name'] = bike_df['End Station Name'].astype(str)
bike_df['End Station ID'] = bike_df['End Station ID'].astype(str)
bike_df['User Type'] = bike_df['User Type'].astype(str)
```

```python
# Display dataframe types
bike_df.dtypes
```

```python
# Drop null values
bike_df = bike_df.dropna()

# Display dataframe
bike_df.head
```

```python
# Save dataframe to csv
bike_df.to_csv('../module18/citibike.csv', index=False)
```
